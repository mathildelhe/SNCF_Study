Others had a longer name in a set than in another (for example: "MONTPELLIER" instead of "MONTPELLIER SAINT-ROCH". Also some mismatch about "ST" / "SAINT" (like ST MALO)

1 station is missing from df_nb_tr: PARIS VAUGIRARD. This station closed in 2019. In our df, it only had data for 2018.

I will manually rename those 24 stations in df to have a perfect match in the formating and be able to proceed to the merging.

## 3.2 Merge the data sets

One last small step before merging: changing the columns names. I will have to do it twice in order to get the info of the number of travelers both for the Departure station and the Arrival station.

```python
# Rename columns for merging on Departure station
df_nb_tr_short.rename(columns={
    'Station name' : 'Departure station',
    'Total travelers 2023': 'Departure station: Total travelers 2023',
    'Total travelers 2022': 'Departure station: Total travelers 2022',
    'Total travelers 2021': 'Departure station: Total travelers 2021',
    'Total travelers 2020': 'Departure station: Total travelers 2020',
    'Total travelers 2019': 'Departure station: Total travelers 2019',
    'Total travelers 2018': 'Departure station: Total travelers 2018',
}, inplace = True)
```

```python
# Merging the 2 data sets. It has to be a left join, to keep rows from intl stations and Paris Vaugirard
df_merged_step1 = df.merge(df_nb_tr_short, on = 'Departure station', how = 'left', indicator = True)
```

```python
# Check result
df_merged_step1
```

I'm expecting some gaps where the departure station is international or Paris Vaugirard. The "Total travelers" columns will say "NaN", because the data was not in the df_nb_tr data set. It's okay; let's keep in mind that those "NaN" mean the data is not available.

```python
# View the rows with missing data
df_merged_step1.loc[df_merged_step1['_merge'] != 'both']
```

```python
# Need to rename "_merge" column to be able to do the 2nd merging
df_merged_step1.rename(columns={'_merge' : '_merge_1'}, inplace = True)
```

```python
# Rename columns for merging on Arrival station
df_nb_tr_short.rename(columns={
    'Departure station' : 'Arrival station',
    'Departure station: Total travelers 2023': 'Arrival station: Total travelers 2023',
    'Departure station: Total travelers 2022': 'Arrival station: Total travelers 2022',
    'Departure station: Total travelers 2021': 'Arrival station: Total travelers 2021',
    'Departure station: Total travelers 2020': 'Arrival station: Total travelers 2020',
    'Departure station: Total travelers 2019': 'Arrival station: Total travelers 2019',
    'Departure station: Total travelers 2018': 'Arrival station: Total travelers 2018',
}, inplace = True)
```

```python
# Merging the 2 data sets. It has to be a left join, to keep rows from intl stations and Paris Vaugirard
df_merged_total = df_merged_step1.merge(df_nb_tr_short, on = 'Arrival station', how = 'left', indicator = True)
```

```python
# Check result
df_merged_total
```
